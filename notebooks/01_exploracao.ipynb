```python
import os
import requests
from dotenv import load_dotenv

# 1. Carrega a chave blindada do seu arquivo .env
load_dotenv()
LASTFM_API_KEY = os.getenv('LASTFM_API_KEY')

# 2. Configura a requisição (usando a URL base oficial)
url = "http://ws.audioscrobbler.com/2.0/"

# O dicionário 'payload' monta os parâmetros da URL de forma limpa e segura
payload = {
    'method': 'artist.getinfo',
    'artist': 'King Crimson', # O nome da banda vai direto aqui!
    'api_key': LASTFM_API_KEY,
    'format': 'json'
}

# 3. Bate na porta da API
response = requests.get(url, params=payload)
data = response.json()

# 4. Exibe os dados extraídos
print("=== PERFIL DA BANDA NO LAST.FM ===")
print(f"Banda: {data['artist']['name']}")

# Pegando os números absolutos e formatando com pontuação de milhar
ouvintes = int(data['artist']['stats']['listeners'])
scrobbles = int(data['artist']['stats']['playcount'])

print(f"Ouvintes totais: {ouvintes:,}".replace(',', '.'))
print(f"Reproduções Globais (Scrobbles): {scrobbles:,}".replace(',', '.'))

print("\n=== TAGS DA COMUNIDADE (GÊNEROS) ===")
# Varre a lista de tags e imprime uma a uma
for tag in data['artist']['tags']['tag']:
    print(f"- {tag['name'].title()}")
```

```python
import pandas as pd

# 1. Tratamento da lista de tags (pegando as 3 principais e unindo com vírgula)
# Isso é crucial porque bancos de dados relacionais odeiam "listas" dentro de uma única célula.
top_tags = [tag['name'].title() for tag in data['artist']['tags']['tag'][:3]]
tags_formatadas = ", ".join(top_tags)

# 2. Estruturando o dicionário que vai virar a primeira linha da tabela
linha_banda = {
    'Banda': [data['artist']['name']],
    'Ouvintes': [int(data['artist']['stats']['listeners'])],
    'Scrobbles': [int(data['artist']['stats']['playcount'])],
    'Top_Generos': [tags_formatadas]
}

# 3. Criando o DataFrame do Pandas (tabela)
df_banda = pd.DataFrame(linha_banda)

# No Jupyter Notebook, não é necessário usar o comando print() para tabelas. 
# Apenas colocar o nome da variável na última linha faz o motor renderizar uma tabela visual limpa.
df_banda
```
